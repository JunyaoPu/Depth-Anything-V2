```python
import cv2
import torch
from depth_anything_v2.dpt import DepthAnythingV2
import matplotlib.pyplot as plt
import numpy as np
```

```python
DEVICE = 'cuda' if torch.cuda.is_available() else 'mps' if torch.backends.mps.is_available() else 'cpu'

model_configs = {
    'vits': {'encoder': 'vits', 'features': 64, 'out_channels': [48, 96, 192, 384]},
    'vitb': {'encoder': 'vitb', 'features': 128, 'out_channels': [96, 192, 384, 768]},
    'vitl': {'encoder': 'vitl', 'features': 256, 'out_channels': [256, 512, 1024, 1024]},
    'vitg': {'encoder': 'vitg', 'features': 384, 'out_channels': [1536, 1536, 1536, 1536]}
}

encoder = 'vitl' # or 'vits', 'vitb', 'vitg'

model = DepthAnythingV2(**model_configs[encoder])
model.load_state_dict(torch.load(f'checkpoints/depth_anything_v2_{encoder}.pth', map_location='cpu'))
model = model.to(DEVICE).eval()

raw_img = cv2.imread('sample/sample.jpg')
depth = model.infer_image(raw_img) # HxW raw depth map in numpy
```

```python
### Plot the normalized depth map

# Assume `depth` is the HxW numpy array containing the depth map
# Normalize the depth map for better visualization
depth_normalized = (depth - np.min(depth)) / (np.max(depth) - np.min(depth))

# Plot the depth map
plt.figure(figsize=(8, 6))
plt.imshow(depth_normalized, cmap='viridis')  # You can also try 'plasma', 'inferno', 'magma', etc.
plt.colorbar(label='Depth')
plt.title('Depth Map Visualization')
plt.axis('off')
plt.show()
```

```python
# Plot the depth map without normalization
plt.figure(figsize=(8, 6))
plt.imshow(depth, cmap='viridis')  # Use 'viridis' or other colormaps like 'plasma', 'inferno', etc.
plt.colorbar(label='Depth')
plt.title('Depth Map Visualization')
plt.axis('off')
plt.show()
```
